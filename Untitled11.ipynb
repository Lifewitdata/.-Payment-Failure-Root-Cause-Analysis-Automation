# Task
Clean the data in the file "transactions.csv" by handling missing values, checking data types, identifying duplicates, analyzing the 'failure reason' column, and validating data consistency.

## Handle missing values

### Subtask:
Identify and address any missing values in the dataset.


**Reasoning**:
Calculate and display the number of missing values for each column to identify where missing values exist.



```python
missing_values = df.isnull().sum()
print("Missing values per column:")
print(missing_values)

# Check for unexpected missing values in columns other than 'failure_reason'
unexpected_missing_columns = missing_values[missing_values > 0].drop('failure_reason', errors='ignore')

if not unexpected_missing_columns.empty:
    print("\nUnexpected missing values found in the following columns:")
    print(unexpected_missing_columns)
else:
    print("\nNo unexpected missing values found in columns other than 'failure_reason'.")
```

## Check data types

### Subtask:
Verify and convert data types as necessary for analysis (e.g., ensure 'timestamp' is datetime).


**Reasoning**:
Inspect the data types of the DataFrame columns and convert the 'timestamp' column to datetime if necessary.



```python
# Inspect current data types
df.info()

# Convert 'timestamp' to datetime if it's not already
if df['timestamp'].dtype != 'datetime64[ns]':
    df['timestamp'] = pd.to_datetime(df['timestamp'])
    print("\n'timestamp' column converted to datetime.")
else:
    print("\n'timestamp' column is already datetime.")

# Re-inspect data types to confirm
df.info()
```

## Identify duplicates

### Subtask:
Detect and decide how to handle duplicate records if any exist.


**Reasoning**:
Calculate and print the number of duplicate rows in the DataFrame.



```python
num_duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {num_duplicates}")

if num_duplicates > 0:
    print("\nInspecting a few duplicate rows:")
    display(df[df.duplicated(keep=False)].head())
else:
    print("\nNo duplicate rows found.")
```

## Analyze 'failure reason'

### Subtask:
Explore the 'failure_reason' column, specifically for failed transactions, and potentially categorize or standardize entries.


**Reasoning**:
Filter the DataFrame to include only failed transactions and then count the occurrences of each unique failure reason.



```python
# Filter for failed transactions
failed_transactions_df = df[df['status'] == 'Failed']

# Count occurrences of each failure reason
failure_reason_counts = failed_transactions_df['failure_reason'].value_counts()

# Print the counts
print("Counts of each failure reason for failed transactions:")
print(failure_reason_counts)
```

## Validate data consistency

### Subtask:
Check for inconsistencies, like 'failure_reason' entries for successful transactions.


**Reasoning**:
Filter the DataFrame to find rows where the status is 'Success' but a failure reason is present, then count these inconsistent records and display the first few if any are found.



```python
# Filter for inconsistencies: 'Success' status with a non-null 'failure_reason'
inconsistent_records = df[(df['status'] == 'Success') & (df['failure_reason'].notnull())]

# Count the number of inconsistent records
num_inconsistencies = len(inconsistent_records)

# Print the number of inconsistencies
print(f"Number of inconsistent records (Success status with failure reason): {num_inconsistencies}")

# If inconsistencies are found, display the first few rows
if num_inconsistencies > 0:
    print("\nFirst 5 inconsistent records:")
    display(inconsistent_records.head())
else:
    print("\nNo inconsistencies found.")
```

Other reasons included 'Bank Unavailable', 'Insufficient Funds', and 'Card Expired', each with similar counts around 2400-2500.

### Insights or Next Steps

*   The data is relatively clean with no unexpected missing values, duplicates, or data type issues requiring immediate correction.
*   The distinct failure reasons are well-defined and do not require standardization for initial analysis.

## Analyze failure distribution

### Subtask:
Count and plot how many transactions failed vs succeeded.


**Reasoning**:
Count the occurrences of each status and plot the distribution using a bar chart.



```python
import matplotlib.pyplot as plt

# Count the occurrences of each unique value in the 'status' column
status_counts = df['status'].value_counts()

# Print the counts
print("Distribution of Transaction Statuses:")
print(status_counts)

# Create a bar plot of the status counts
plt.figure(figsize=(8, 6))
status_counts.plot(kind='bar', color=['green', 'red'])
plt.title('Distribution of Transaction Statuses')
plt.xlabel('Status')
plt.ylabel('Number of Transactions')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

## Visualize trend of failures over time

### Subtask:
Visualize the trend of failures over time (monthly or weekly).


**Reasoning**:
Resample the DataFrame by month, count failed transactions, and plot the monthly trend.



```python
# Resample the DataFrame by month and count the number of 'Failed' transactions
monthly_failed_transactions = df[df['status'] == 'Failed'].resample('M', on='timestamp').size()

# Create a line plot of the monthly failed transaction counts
plt.figure(figsize=(10, 6))
monthly_failed_transactions.plot(kind='line', marker='o')
plt.title('Monthly Trend of Failed Transactions')
plt.xlabel('Month')
plt.ylabel('Number of Failed Transactions')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Root cause breakdown

### Subtask:
Analyze top failure reasons and their proportions, and plot a bar chart or pie chart of the most frequent failure causes.


**Reasoning**:
Calculate the proportion of each failure reason and print them, then create and display a bar plot of the failure reason counts.



```python
# Calculate the proportion of each failure reason
total_failed_transactions = len(failed_transactions_df)
failure_reason_proportions = failure_reason_counts / total_failed_transactions

# Print the proportions
print("\nProportion of each failure reason for failed transactions:")
print(failure_reason_proportions)

# Create a bar plot of the failure reason counts
plt.figure(figsize=(10, 7))
failure_reason_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Failure Reasons for Failed Transactions')
plt.xlabel('Failure Reason')
plt.ylabel('Number of Transactions')
plt.xticks(rotation=45, ha='right') # Rotate labels for better readability
plt.tight_layout()
plt.show()
```

## Failure rate by payment method

### Subtask:
Calculate and visualize the failure rate for each payment method.


**Reasoning**:
Group the data by payment method and status, unstack the result, calculate the failure rate, and then plot the failure rates for each payment method.



```python
# Group by payment_method and status, then count
payment_status_counts = df.groupby(['payment_method', 'status']).size().unstack(fill_value=0)

# Calculate failure rate
# Handle cases where a payment method might not have any transactions (denominator would be zero)
payment_status_counts['total'] = payment_status_counts['Failed'] + payment_status_counts['Success']
payment_status_counts['failure_rate'] = payment_status_counts['Failed'] / payment_status_counts['total']

# Print failure rates
print("Failure rate by Payment Method:")
print(payment_status_counts['failure_rate'])

# Plot failure rates
plt.figure(figsize=(10, 6))
payment_status_counts['failure_rate'].plot(kind='bar', color='orange')
plt.title('Failure Rate by Payment Method')
plt.xlabel('Payment Method')
plt.ylabel('Failure Rate')
plt.xticks(rotation=45, ha='right')
plt.ylim(0, 1) # Set y-axis limit between 0 and 1 for rate
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

## Time-based analysis

### Subtask:
Detect if failures spike during specific hours of the day or days of the week.


**Reasoning**:
Extract the hour and day of the week from the timestamp for failed transactions, then group by hour and day to count failures and prepare for plotting.



```python
# Extract hour of the day and day of the week for failed transactions
failed_transactions_df['hour_of_day'] = failed_transactions_df['timestamp'].dt.hour
failed_transactions_df['day_of_week'] = failed_transactions_df['timestamp'].dt.day_name()

# Group by hour of the day and count failures
failures_by_hour = failed_transactions_df.groupby('hour_of_day').size()

# Group by day of the week and count failures
# Ensure correct order for days of the week
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
failures_by_day = failed_transactions_df.groupby('day_of_week').size().reindex(days_order)

# Plot failures by hour
plt.figure(figsize=(12, 6))
failures_by_hour.plot(kind='bar', color='purple')
plt.title('Number of Failed Transactions by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Failed Transactions')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# Plot failures by day of the week
plt.figure(figsize=(10, 6))
failures_by_day.plot(kind='bar', color='teal')
plt.title('Number of Failed Transactions by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Number of Failed Transactions')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
